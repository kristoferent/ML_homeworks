```python
import math 
def dist(a, b, f1, f2, coef1, coef2):
    return math.sqrt(((a[f1] - b[f1])*coef1)**2 + ((a[f2] - b[f2])*coef2)**2)
```

```python
def knn(test_df, train_df, f1, f2, k, coef1, coef2):
    predictions = pd.DataFrame([[0]], columns = ['pred'])
    for i in range(len(test_df)):
        all_points = [[dist(test_df.loc[i], train_df.loc[j], f1, f2, coef1, coef2), train_df.loc[j]['SeriousDlqin2yrs']]  for j in range(len(train_df))]
        k_nearest = [0, 0]
        for point in (sorted(all_points)[:k]):
            k_nearest[int(point[1])] += 1
        if k_nearest[0] > k_nearest[1]:
            predictions.loc[i] = 0
        else:
            predictions.loc[i] = 1
    return predictions
```

```python
feature1 = 'Debt'
feature2 = 'MonthlyIncome'
coef1 = 1.0
coef2 = 1.0
k = 2
df_test = df.head(0)
df_test = df_test.append(df.tail(100), ignore_index=True)
df_train = df.head(100)
res = knn(df_test, df_train, feature1, feature2, k, coef1, coef2)
print(res)
```

```python
plt.scatter(df_test[df_test['SeriousDlqin2yrs']==0][feature1], df_test[df_test['SeriousDlqin2yrs']==0][feature2], color='blue')
plt.scatter(df_test[df_test['SeriousDlqin2yrs']==1][feature1], df_test[df_test['SeriousDlqin2yrs']==1][feature2], color='red')
```
